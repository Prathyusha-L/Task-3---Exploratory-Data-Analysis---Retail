### Data Visualisation

```python
sns.pairplot(sales)
```

```python
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.displot(sales['Sales'],ax=axes[0,0])
sns.displot(sales['Profit'],ax=axes[0,1])
sns.displot(sales['Discount'],ax=axes[1,0])
sns.displot(sales['Quantity'],ax=axes[1,1])
plt.show()
```

```python
fig,ax=plt.subplots(1,1,figsize=(12,7))
sns.countplot(sales['Quantity'],hue=sales['Region'])
plt.show()
```
